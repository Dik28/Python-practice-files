```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
iris= sns.load_dataset("iris")
iris.head()
```

```python
fmri= sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
#Changing the Aesthetics
# adding a hue

sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
# Adding style
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event", style="event")
plt.show()
```

```python
#Adding markers
sns.lineplot(x="timepoint", y="signal",hue="event",style="event", markers=True, data=fmri)
plt.show()
```
